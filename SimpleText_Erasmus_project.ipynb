```python
import pandas as pd
from google.colab import drive
import requests
import re

drive.mount('/content/drive')
test_s=pd.read_json('/content/drive/MyDrive/Erasmus/brest/test/simpletext-task2-test-small.json')
test_10 = test_s[:10]
API_URL = "https://api-inference.huggingface.co/models/bigscience/bloom"
headers = {"Authorization": f"Bearer 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"}
def query(payload):
	response = requests.post(API_URL, headers=headers, json=payload)
	return response.json()
test_10
```

```python
# 1. Find other difficult scientific terms

prompt = "\
Sentence: Penetration tests have become a valuable tool in the cyber security defence strategy in terms of detecting vulnerabilities.\n\
Difficult scientific terms: penetration tests, cyber security, defense strategy, vulnerabilities [END]\n\
\n\
Sentence: Treating search engines as editorial products with intrinsic biases can help understand the structure of information flows in new media.\n\
Difficult scientific terms: editorial products, intrinstic biases, structure of information flows, new media [END]\n\
\n\
Sentence: The immune-based algorithm Artificial Immune Recognition System (AIRS) is applied to Duke Energy distribution systems outage data and we investigate its capability to classify imbalanced data.\n\
Difficult scientific terms: immune-based algorithm, Artificial Immune Recognition System (AIRS), Duke Energy distribution systems, outage data, imbalanced data [END]\n\
\n\
Sentence: Everyone wants to have more money and be happy.\n\
Difficult scientific terms: [END]"

def generate_from_prompt(prompt,snt):
  try:
    p=prompt+'Sentence: '+snt+'\nDifficult scientific terms:'
    output = query({
    "inputs": p
    })
    s=re.split('\nDifficult scientific terms:',output[0]['generated_text'],re.DOTALL)[-1]
    return s
  except Exception as e:
    return e

def cut_string(text, index):
  if index != -1:
    return text[:(index-1)]
  else:
    return text
```

```python
test_10['difficult_scientific_terms']=test_10['source_snt'].apply(lambda x:generate_from_prompt(prompt,x))
test_10
```
